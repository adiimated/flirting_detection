### Merging the 4 Datasets

```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
flirting = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/flirting detection/cleaning/flirting_cleaned.csv")
flirting_rated = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/flirting detection/cleaning/flirting_rated_cleaned.csv")
test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/flirting detection/cleaning/test_cleaned.csv")
validation = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/flirting detection/cleaning/validation_cleaned.csv")
```

```python
merged_df = pd.concat([flirting, flirting_rated, test, validation], ignore_index=True)
```

```python
merged_df.head()
```

```python
# Save the merged DataFrame to a new CSV file
output_path = "/content/drive/MyDrive/Colab Notebooks/flirting detection/text_detection/data.csv"
merged_df.to_csv(output_path, index=False)
```
